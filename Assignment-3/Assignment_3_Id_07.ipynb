```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

**Assignment 1: Basic Matrix Operations**

```python

#Creating two 3Ã—3 matrices with random integers in Python.
A = np.random.randint(0, 20, size=(3, 3))
B = np.random.randint(0, 20, size=(3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)


# Performing addition, subtraction, and multiplication operations on these matrices.
matrix_sum = A + B
matrix_diff = A - B
matrix_mul = A * B
matrix_prod = np.dot(A, B)

print("Matrix Sum:\n", matrix_sum)
print("Matrix Difference:\n", matrix_diff)
print("Matrix Product:\n", matrix_prod)
print("Matrix Multiplication:\n", matrix_mul)



# Computing the determinant, inverse, and rank of one of the matrices.
det_A = np.linalg.det(A)
print("detrminate matrix of f : \n",det_A,"\n")

if np.linalg.det(A) != 0:
     inv_A = np.linalg.inv(A)
     print("inverse matrix of f : \n",inv_A,"\n")
else:
     print("f is not invertible")

rank_A = np.linalg.matrix_rank(A)
print("rank of a : \n",rank_A,"\n")
```

***1. Generate and Plot Two Random Vectors:***

```python
# a. Creating two vectors, each containing 15 random floats of my choice.

A = np.random.randint(1, 15, (3, 3))
B = np.random.randint(1, 15, (3, 3))

print(A)
print(B)

A_flat = A.flatten()
B_flat = B.flatten()

x = np.arange(len(A_flat))
print(x)

# b. Ploting these vectors on the same graph using matplotlib to compare how they look.
plt.plot(x, A_flat, marker='o', linestyle='-', label="Matrix A")
plt.plot(x, B_flat, marker='s', linestyle='--', label="Matrix B")

# c. Labeling the axes appropriately and give the plot a title.
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Flattened Matrices A and B")
plt.legend()
plt.grid(True)


plt.show()
```

***2. Generate and Visualize a Random Matrix:***

```python
#a. Creating a 4x4 matrix with random values of my choice.
a = np.array([[1, 2, -9, 11], [2, -1, 2, 22], [3, -4, 3, 9],[1,9,8,7]])
b = np.array([[1, 2, 3, 33], [4, 5, 6, 66], [7, 8, 9, 99],[2,9,8,7]])


#b. Display the matrix using a heatmap with appropriate color mapping (e.g.,seaborn or matplotlib). & c. Add labels to the heatmap to indicate row and column numbers.
sns.heatmap(a, annot=True, cmap='coolwarm', linewidths=0.5, linecolor ='black')
plt.title('Heatmap of 3x3 Matrix')
plt.show()

sns.heatmap(b, annot=True, cmap='coolwarm', linewidths=0.5, linecolor ='black')
plt.title('Heatmap of 3x3 Matrix')
plt.show()
```

***3. Matrix Operations and Visualization:***

```python
#a. Creating a 4x4 matrix with random values of my choice.
A = np.array([[1, 2, -9, 11], [2, -1, 2, 22], [3, -4, 3, 9],[1,9,8,7]])
B = np.array([[1, 2, 3, 33], [4, 5, 6, 66], [7, 8, 9, 99],[2,9,8,7]])


print("Matrix A:\n", A)
print("Matrix B:\n", B)

# b. Performing the following operations: i. Matrix Addition, ii. Matrix Subtraction, iii. Matrix Multiplication (ensure matrices are compatible for multiplication)
matrix_sum = A + B
matrix_diff = A - B
matrix_prod = np.dot(A, B)

# c. For each of these operations, visualize the result using bar plots. Each bar represented one element of the matrix.
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Sum
axes[0].bar(range(1, 17), matrix_sum.flatten(), color='b')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Difference
axes[1].bar(range(1, 17), matrix_diff.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Product
axes[2].bar(range(1, 17), matrix_prod.flatten(), color='r')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
